```python
import matplotlib.pyplot as plt


#empty lists store iterations and resolutions
iterations = []
current_resolutions = []

#log file, extracts iterations and their resolutions
with open("j82.out", "r") as file:
    for line in file:
        if "Expectation iteration" in line:
            iteration = int(line.split()[2])
            iterations.append(iteration)
        elif "CurrentResolution=" in line:
            resolution = float(line.split()[1])
            current_resolutions.append(resolution)

#plot
plt.figure(figsize=(10, 6))
plt.plot(iterations, current_resolutions, marker='o', linestyle='-', color='y')
plt.title("j82 3D classification: Resolution development throughout iterations ")
plt.xlabel("Iteration")
plt.ylabel("Resolution (Angstroms)")
plt.xticks(range(min(iterations), max(iterations), 2))
plt.grid(True)
plt.tight_layout()
plt.show()
```
