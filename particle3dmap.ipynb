```python
import pandas as pd
import matplotlib.pyplot as plt
filtered_data = pd.read_csv('tbl_filtered_bin4.tbl', sep='\s+', header=None)
#extracts coordinates and cc values
x_filtered = filtered_data[23]
y_filtered = filtered_data[24]
z_filtered = filtered_data[25]
cc_filtered = filtered_data[9]

#to find min and max cc values
#value_column_9 = filtered_data[9].min()
#value_column_9

#creates 3d scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
#colour adjustment
scatter = ax.scatter(x_filtered, y_filtered, z_filtered, c=cc_filtered, cmap="magma", s= 0.5, alpha=0.4, vmin=0.4,vmax=0.77)

#axis, labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Spatial distribution of particles and their CC values')

#colorbar
cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('CC Value')
plt.show()


```
