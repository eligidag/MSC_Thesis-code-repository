```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

filtered_data = pd.read_csv('tbl_filtered_bin4.tbl', sep='\s+', header=None)

#coordinates
x_filtered = filtered_data[23]
y_filtered = filtered_data[24]
z_filtered = filtered_data[25]
cc_filtered = filtered_data[9]

#figure with individual subplots for each distribution
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

#x distribution
sns.histplot(x_filtered, ax=axes[0, 0], kde=True, color='blue', bins=50)
axes[0, 0].set_title('Distribution of particles within X axis')
axes[0, 0].set_xlabel('X coordinate')
axes[0, 0].set_ylabel('Particle frequency')
axes[0, 0].grid(True)
axes[0, 0].set_axisbelow(True)


#y distribution
sns.histplot(y_filtered, ax=axes[0, 1], kde=True, color='green', bins=50)
axes[0, 1].set_title('Distribution of particles within Y axis')
axes[0, 1].set_xlabel('Y coordinate')
axes[0, 1].set_ylabel('Particle Count')
axes[0, 1].grid(True)
axes[0, 1].set_axisbelow(True)


#z distribution
sns.histplot(z_filtered, ax=axes[1, 0], kde=True, color='red', bins=50)
axes[1, 0].set_title('Distribution of particles within Z axis')
axes[1, 0].set_xlabel('Z coordinate')
axes[1, 0].set_ylabel('Particle Count')
axes[1, 0].grid(True)
axes[1, 0].set_axisbelow(True)

#cc distribution
sns.histplot(cc_filtered, ax=axes[1, 1], kde=True, color='purple', bins=50)
axes[1, 1].set_title('Distribution of CC values')
axes[1, 1].set_xlabel('CC Value')
axes[1, 1].set_ylabel('Particle Count')
axes[1, 1].grid(True)
axes[1, 1].set_axisbelow(True)


plt.tight_layout()
plt.savefig("particle_coord_distr.png")
plt.show()
```
